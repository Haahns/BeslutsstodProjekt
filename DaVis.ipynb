```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('PhonesNew.csv', encoding='UTF-8', on_bad_lines='skip')
df.head(2)
```

```python
# Create a histogram/bar chart of Weight with 33-gram bins
import numpy as np

# Get the weight data
weights = df['Mobile Weight']

# Create bins with 33-gram intervals
min_weight = weights.min()
max_weight = weights.max()
bin_size_weight = 5
bins = np.arange(min_weight, max_weight + 5, bin_size_weight) # 33-gram bins

# Create the histogram
plt.figure(figsize=(12, 6))
plt.hist(weights, bins=bins, edgecolor='black', alpha=0.7, color='red')
plt.xlabel('Weight (g)')
plt.ylabel('Amount')
plt.title(f'Weight distribution, bin size: {bin_size_weight}')
plt.grid(True, alpha=0.3)

# Add some statistics
print(f"Weight range: {min_weight}g - {max_weight}g")
print(f"Number of bins: {len(bins)-1}")
print(f"Bin width: 33 grams")

#plt.tight_layout()
#plt.show()
```

```python
# Option 1: Exclude the dominant weight range to see other patterns better
weights = df['Mobile Weight']
  # Gets 'RAM'

# Find which bin has the most devices
counts, bin_edges = np.histogram(weights, bins=np.arange(weights.min(), weights.max() + 33, 33))
max_count_index = np.argmax(counts)
dominant_range = f"{bin_edges[max_count_index]:.0f}-{bin_edges[max_count_index+1]:.0f}g"

print(f"Dominant weight range: {dominant_range} with {counts[max_count_index]} devices")

# Create filtered data excluding the dominant range
filtered_weights = weights[(weights < bin_edges[max_count_index]) | (weights >= bin_edges[max_count_index+1])]

print(f"Original devices: {len(weights)}")
print(f"After filtering out dominant range: {len(filtered_weights)}")

# Create new bins for filtered data
min_filtered = filtered_weights.min()
max_filtered = filtered_weights.max()
filtered_bins = np.arange(min_filtered, max_filtered + 33, 33)

# Plot the filtered histogram
plt.figure(figsize=(12, 6))
plt.hist(filtered_weights, bins=filtered_bins, edgecolor='black', alpha=0.7, color='lightcoral')
plt.xlabel('Weight (grams)')
plt.ylabel('Number of Devices')
plt.title(f'Device Weight Distribution (Excluding {dominant_range})')
plt.grid(True, alpha=0.3)

plt.tight_layout()

plt.show()
plt.show()
```

```python
#Här visas både X-axelns värde i botten OCH Y-värdet ovanpå varje stapel

# Create a histogram/bar chart of data with 33-gram bins
import numpy as np

# Get the data data
#data = df['Mobile data']
#data = df['RAM']
#data = df['Battery Capacity']
#data = df['Screen size']
#data = df['Front Camera']
data = df['Back Camera'] #ok

column_name = data.name  # Gets 'RAM'

# Create bins with 33-gram intervals
min_data = data.min()
max_data = data.max()
bin_size = 1
bins = np.arange(min_data, max_data + 1, bin_size) 

# Create the histogram
plt.figure(figsize=(16, 6))
counts, bin_edges, patches = plt.hist(data, bins=bins, edgecolor='black', alpha=0.7, color='skyblue')

# Find positions where bars exist and set custom x-axis ticks
bar_positions = []
bar_labels = []
for i, count in enumerate(counts):
    if count > 0:
        bin_center = (bin_edges[i] + bin_edges[i+1]) / 2
        bar_positions.append(bin_center)
        #bar_labels.append(f"{bin_edges[i]:.0f}")
        bar_labels.append(f"{bin_center:.0f}")

# Set x-axis ticks to only show values where there are bars
plt.xticks(bar_positions, bar_labels)

# Add Y values (count) on top of each bar
for i, count in enumerate(counts):
    if count > 0:
        bin_center = (bin_edges[i] + bin_edges[i+1]) / 2
        plt.text(bin_center, count + 0.5, str(int(count)), 
                ha='center', va='bottom', fontweight='light')

plt.xlabel('Range')
plt.ylabel('Amount')
plt.title(f"{column_name} : {min_data}MP - {max_data}MP, bin size:{bin_size}")
plt.grid(True, alpha=0.3)

# Add some statistics
print(f"{column_name} range: {min_data} - {max_data}")
print(f"Number of bins: {len(bins)-1}")
print(f"Bin width: {bin_size}")

plt.tight_layout()
plt.show()
```

```python
# Static distance between bars, excluding empty slots completely
import numpy as np

# Get the data
# Get the data data
#data = df['Mobile data']
#data = df['RAM']
#data = df['Battery Capacity']
data = df['Screen size']
#data = df['Front Camera']
#data = df['Back Camera'] #ok
column_name = data.name

min_data = data.min()
max_data = data.max()

# Get unique values and their counts (this excludes empty slots)
unique_values = sorted(data.unique())
counts = [sum(data == value) for value in unique_values]

# Create bar positions with static spacing (ignoring actual data values)
bar_positions = list(range(len(unique_values)))  # 0, 1, 2, 3, 4...
bar_width = 0.5  # Width of each bar

# Create the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(bar_positions, counts, width=bar_width, 
               edgecolor='black', alpha=0.7, color='blue')

# Set custom x-axis labels to show actual data values
plt.xticks(bar_positions, [f"{val:.0f}" for val in unique_values])

# Add Y values on top of each bar
for i, (pos, count) in enumerate(zip(bar_positions, counts)):
    plt.text(pos, count + 0.5, str(int(count)), 
            ha='center', va='bottom', fontweight='light')

plt.xlabel('Values')
plt.ylabel('Amount')
plt.title(f"{column_name} : {min_data}MP - {max_data}MP, bin size:{bin_size}")
plt.grid(True, alpha=0.3)

# Add statistics
print(f"Unique {column_name} values: {unique_values}")
print(f"Number of unique values: {len(unique_values)}")
print(f"Total devices: {sum(counts)}")

plt.tight_layout()
plt.show()

# Show comparison: actual vs evenly spaced positions
# print(f"\nActual data values: {unique_values}")
# print(f"Evenly spaced positions: {bar_positions}")
# print("Each bar is exactly 1 unit apart, regardless of actual value gaps")
```
