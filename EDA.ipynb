```python
# Count unique values in DataFrame columns

# Method 1: .nunique() - returns the count of unique values
print("Number of unique companies:", df['Company Name'].nunique())
print("Number of unique RAM values:", df['RAM'].nunique())
print("Number of unique screen sizes:", df['Screen size'].nunique())

print("\n" + "="*50 + "\n")

# Method 2: .value_counts() - shows each unique value and its frequency
print("Company distribution:")
print(df['Company Name'].value_counts())

print("\nRAM distribution:")
print(df['RAM'].value_counts())

print("\n" + "="*50 + "\n")

# Method 3: .unique() - returns array of unique values
print("Unique RAM values:", sorted(df['RAM'].unique()))
print("Unique screen sizes:", sorted(df['Screen size'].unique()))

print("\n" + "="*50 + "\n")

# Method 4: For all columns at once
print("Unique value counts for all columns:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
```

```python
# Print all Apple model names

# Method 1: Basic filtering
apple_models = df[df["Company Name"] == "Apple"]["Model Name"]
print("Apple models:")
for model in apple_models:
    print(f"- {model}")

print("\n" + "="*50 + "\n")

# Method 2: More concise - direct print
print("Apple models (direct print):")
print(df[df["Company Name"] == "Apple"]["Model Name"].tolist())

print("\n" + "="*50 + "\n")

# Method 3: Get unique Apple models (remove duplicates)
unique_apple_models = df[df["Company Name"] == "Apple"]["Model Name"].unique()
print(f"Unique Apple models ({len(unique_apple_models)} total):")
for model in sorted(unique_apple_models):
    print(f"- {model}")

print("\n" + "="*50 + "\n")

# Method 4: Count how many times each Apple model appears
print("Apple model counts:")
apple_counts = df[df["Company Name"] == "Apple"]["Model Name"].value_counts()
print(apple_counts)
```

```python
# Create pie chart from Company Name value_counts

# Get value counts
company_counts = df['Company Name'].value_counts()

# Alternative: Show only top companies, group others as "Others"
top_n = 13  # Show top 6 companies
top_companies = company_counts.head(top_n)
others_count = company_counts.tail(len(company_counts) - top_n).sum()

# Create labels and sizes for top companies + others
labels_grouped = top_companies.index.tolist() + ['Övriga']
sizes_grouped = top_companies.values.tolist() + [others_count]

#print(f"Top {top_n} companies + Others:")
#print("Labels:", labels_grouped)
#print("Sizes:", sizes_grouped)

# Create grouped pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes_grouped, labels=labels_grouped, autopct='%1.1f%%', startangle=90)
plt.title(f'Företag')
plt.axis('equal')
plt.show()
```

```python
# Create pie chart from Company Name value_counts

# Get value counts
company_counts = df["Launched Year"].value_counts()

# Alternative: Show only top companies, group others as "Others"
top_n = 20  # Show top 6 companies
top_companies = company_counts.head(top_n)
others_count = company_counts.tail(len(company_counts) - top_n).sum()

# Create labels and sizes for top companies + others
labels_grouped = top_companies.index.tolist() + ['Others']
sizes_grouped = top_companies.values.tolist() + [others_count]

#print(f"Top {top_n} companies + Others:")
#print("Labels:", labels_grouped)
#print("Sizes:", sizes_grouped)

# Create grouped pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes_grouped, labels=labels_grouped, autopct='%1.1f%%', startangle=270)
plt.title(f'Companies')
plt.axis('equal')
plt.show()
```
